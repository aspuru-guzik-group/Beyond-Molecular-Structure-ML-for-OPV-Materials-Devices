```python
import pandas as pd
from pathlib import Path
from rdkit import Chem

desktop = Path.home() / "Desktop"
saeki = pd.read_csv("Saeki_corrected.csv")
```

```python
downloads = Path.home() / "Downloads"
dataset = pd.read_excel(downloads / "FINAL Machine Learning OPV Parameters.xlsx")
```

```python
dataset.head()
```

```python
acceptors = pd.read_csv(desktop / "acceptors.csv", index_col="Acceptor")
donors = pd.read_csv(desktop / "donors.csv", index_col="Donor")

acceptors['SMILES'] = [Chem.CanonSmiles(smi) for smi in acceptors['SMILES']]
donors['SMILES'] = [Chem.CanonSmiles(smi) for smi in donors['SMILES']]

acceptors.head()
donors.head()
```

```python
d_list = []
for i, donor in enumerate(dataset["Donor Molecule"]):
    d_list.append(donors.at[donor, "SMILES"])

a_list = []
for i, acceptor in enumerate(dataset["Acceptor Molecule"]):
    a_list.append(acceptors.at[acceptor, "SMILES"])

dataset["D SMILES"] = d_list
dataset["A SMILES"] = a_list
dataset.head()
```

```python
da_list = []
for i in range(len(dataset)):
    d = dataset.at[i, "D SMILES"]
    a = dataset.at[i, "A SMILES"]
    da_list.append(f"{d}.{a}")
dataset["DonorAcceptor"] = da_list
dataset.head()
```

```python
saeki.to_csv((desktop / "Saeki_corrected.csv"), index=False)
```

```python
def convert_verify_smiles(column: str):
    unkown = 0
    b_zn = 0
    for i, smi in enumerate(saeki[column]):
        if ("B" in smi) or ("Zn" in smi):
            b_zn += 1
        else:
            try:
                new_smi = Chem.CanonSmiles(smi)
                saeki.at[i, column] = new_smi
            except:
                unkown += 1
                print(smi)
                corr_smi = input("Enter correct SMILES")
                new_smi = Chem.CanonSmiles(corr_smi)
                saeki.at[i, column] = new_smi
    print(unkown)
    print(b_zn)
```

```python
convert_verify_smiles("n(SMILES)")
```

```python
convert_verify_smiles("p(SMILES)")
```

```python
da_list = []
for i in range(len(saeki)):
    d = saeki.at[i, "p(SMILES)"]
    a = saeki.at[i, "n(SMILES)"]
    da_list.append(f"{d}.{a}")
saeki["DonorAcceptor"] = da_list
```

```python
merged = None
merged = dataset.merge(saeki, left_on=[
                                       "DonorAcceptor",
                                       "PCE (%)",
                                       # "Voc (V)", "Jsc (mA cm^-2)"
                                       ],
                              right_on=[
                                        "DonorAcceptor",
                                        "PCE_max(%)",
                                        # "Voc(V)", "Jsc(mA/cm2)"
                                        ], indicator=True)
print(len(merged))
merged.head()
```

```python
print(len(dataset), len(saeki), len(merged))
```

```python
merged.to_csv(desktop / 'merged_da_pce.csv', index=False)
```
